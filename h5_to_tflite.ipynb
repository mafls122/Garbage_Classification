```python
from tensorflow.keras.models import load_model 

model = load_model('/content/drive/My Drive/gdataset/efficientnetB1.h5')
```

```python
model.summary()
```

```python
import tensorflow

converter = tensorflow.lite.TFLiteConverter.from_keras_model(model)

tflite_model = converter.convert()
open('efficientnetB1.tflite', 'wb').write(tflite_model)
```
